## Quantity and type of the attributes

```python
def dataset_stats(df):
    numerical_cols = 0
    categorical_cols = 0
    types = df.dtypes
    if len(types[types == object]) > 0:
        print("Cardinalities :")
    max_cardinality = 0
    for col, col_type in zip(df, types):
        if str(col_type) == 'int64' or str(col_type) == 'float64' :
            numerical_cols += 1
        elif col_type == object:
            categorical_cols += 1
            cardinality = len(pd.Index(df[col]).value_counts())
            print('  ', str(df[col].name) + ": " + str(cardinality))
            if cardinality > max_cardinality :
                max_cardinality = cardinality
    print("Numerical attributes :", numerical_cols)
    print("Categorical attributes :", categorical_cols)
    print("Highest cardinality :", max_cardinality)
```

```python
for name, df in datasets_dict.items() :
    print("\n====================")
    print(name)
    print("Shape =", df.shape)
    dataset_stats(df)
```
